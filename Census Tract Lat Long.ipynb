```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
```

```python
#table = BeautifulSoup(open("./Data/CensusLatLong.html", 'r').read())
table = pd.read_html("./Data/CensusLatLong2.html", flavor='bs4')
table
```

```python
table = table[0]
table.dtypes
```

```python
df = table[["GEOID","COUNTY", "TRACT", "CENTLAT", "CENTLON"]]
df = df.loc[df["COUNTY"] == 71]
df["GEOID"] = df["GEOID"].astype(str)
df["GEOID"] = df["GEOID"].apply(lambda x: x[:11])
df["GEOID"] = df["GEOID"].astype(np.int64)
df["TRACT"] = df["TRACT"].astype(np.float64)
df.head()
```

```python
df.to_csv("./Data/CensusTractToLatLong2.csv")
```

```python
df = df.groupby("GEOID")["CENTLAT", "CENTLON"].mean()
display(df)
```

```python
race_df = pd.read_csv("./Data/cleaned_race_data.csv", index_col=0)
race_df.dtypes
```

```python
race_df["Geo_ID"] = race_df["Geo_ID"].apply(lambda x: x[9:])
race_df["Geo_ID"] = race_df["Geo_ID"].astype(np.int64)
```

```python
print(df.loc[22071006000])
new_df = race_df.merge(df, how="left", left_on="Geo_ID", right_on="GEOID")
```

```python
income_df = pd.read_csv("./Data/cleaned_income_data.csv", index_col=0)
whole_df = new_df.merge(income_df, on=["Census Tract", "Year"], how="inner")
```
