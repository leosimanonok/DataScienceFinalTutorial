```python
import geopandas as gpd
import pandas as pd
import numpy as np
import folium as fl
from pandas.io.json import json_normalize
import matplotlib.pyplot as plt
import json
```

```python
geo_df = gpd.read_file('./Data/tract_nola.geojson')
geo_df
```

```python
data = pd.read_csv("./Data/cleaned_race_data.csv")
latLong = pd.read_csv('./Data/CensusTractToLatLong.csv')
display(data)
display(latLong)
```

```python
data["Percent White"] = data["White"] / data["Total Population"]
data["Percent Black"] = data["Black"] / data["Total Population"]
data = data.merge(latLong, how="left", left_on="Geo_ID", right_on="GEOID")
display(data)
```

```python
data_2010 = data.loc[data["Year"] == 2010]
data_2018 = data.loc[data["Year"] == 2018]
```

```python
data_2010.set_index("Geo_ID", inplace=True)
data_2018.set_index("Geo_ID", inplace=True)
```

```python
change_white = data_2018["Percent White"] - data_2010["Percent White"]
```

```python
latLong
```

```python
latLong = latLong.merge(change_white, how="inner", left_on="GEOID", right_on="Geo_ID")
latLong
```

```python
latLong.rename(columns={"Percent White" : "Change in Percent White"}, inplace=True)
```

```python
latLong["GEOID"] = latLong["GEOID"].astype(str)
```

```python
latLong
```

```python
m = fl.Map(
    location = [29.9511, -90.0715],
    zoom_start = 10,
    tiles="Stamen Toner")
```

```python
m
```

```python
fl.Choropleth(
    geo_data="./Data/tract_nola.geojson",
    name="Change in Percent White from 2010 to 2018 by Census Tract",
    data=latLong,
    columns=["GEOID", "Change in Percent White"],
    key_on='feature.properties.GEOID', 
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="Change in Percent White from 2010 to 2018 by Census Tract"
    ).add_to(m)
```

```python
m
```

```python
data_2010_markers = data_2010.loc[(data_2010["Percent Black"] < 0.6) & (data_2010["Percent Black"] > 0.4)]
```

```python
data_2010_markers.apply(lambda x: fl.Circle([x["CENTLAT"], x["CENTLON"]], radius=10).add_to(m), axis=1)
```

```python
m
```

```python
data_2010_markers.dtypes
data_2010_markers["GEOID"] = data_2010_markers["GEOID"].astype(str)
print(data_2010_markers.dtypes)
print(latLong.dtypes)
```

```python
data_2010_markers = data_2010_markers.merge(latLong, how='left', on="GEOID")
```

```python
data = pd.read_csv('./Data/cleaned_income_data.csv')
```

```python
data.head()
```

```python
new_data = data.loc[data["Year"] == 2010]
data_2011 = data.loc[data["Year"] == 2011]
new_data = new_data.set_index("Geo_ID")
data_2011 = data_2011.set_index("Geo_ID")
new_data["Mean Income / Median Income"] = new_data["Household mean income (dollars)"] / new_data["Household median income (dollars)"]
#new_data["Change in Median Income"] = new_data["Household median income (dollars)"] - data_2011["Household median income (dollars)"]
#new_data["Change in Mean Income"] = new_data["Household mean income (dollars)"] - data_2011["Household mean income (dollars)"]
#new_data["Change Mean / Change Median"] = new_data["Change in Mean Income"] / new_data["Change in Median Income"]
new_data.head()
```

```python
for i in range(2011, 2018):
    
    data_year = data.loc[data["Year"] == i]
    data_next = data.loc[data["Year"] == i+1]

    data_year = data_year.set_index("Geo_ID")
    data_next = data_next.set_index("Geo_ID")
    data_year["Mean Income / Median Income"] = data_year["Household mean income (dollars)"] / data_year["Household median income (dollars)"]
    #data_year["Change in Median Income"] = data_year["Household median income (dollars)"] - data_next["Household median income (dollars)"]
    #data_year["Change in Mean Income"] = data_year["Household mean income (dollars)"] - data_next["Household mean income (dollars)"]
    #data_year["Change Mean / Change Median"] = data_year["Change in Mean Income"] / data_year["Change in Median Income"]
    new_data = pd.concat([new_data, data_year])

new_data
```

```python
data_2010_markers
```

```python
data_2010_markers = data_2010_markers.loc[data_2010_markers["Change in Percent White"] > 0]
data_2010_markers["GEOID"] = data_2010_markers["GEOID"].astype(np.int64)
data_2010_markers.dtypes
```

```python
new_data = new_data.merge(data_2010_markers, how='inner', left_on="Geo_ID", right_on="GEOID")
new_data
```

```python
new_data["GEOID"].unique()
```

```python
fig, ax = plt.subplots(figsize = (30,10))
year = np.linspace(2010, 2017, 8)

for GEOID in new_data["GEOID"].unique():
    ax.scatter(year, new_data.loc[new_data["GEOID"] == GEOID]["Household mean income (dollars)"], color = "black")


ax.plot(year, new_data[["Household mean income (dollars)", "Year_x"]].groupby(["Year_x"]).mean(), marker = "D", color="orange", label = "Average")
#ax.set_ylim([0, 5])
#ax.plot(year, np.zeros(8), color="blue")
#ax.legend()
```
